```python
#В этом графе(выше) можно видеть что с повышением возраста в среднем
#повышается и плата, которую люди отдают за все приколы. Это линейная
#зависимость т.е. charges = w(slope)*age+b(bias)
```

```python
def estimate_charges(age,w,b):
    return w * age + b

def rmse(targets, predictions):
    return np.sqrt(np.mean(np.square(targets - predictions)))

def try_parameters(w, b):
    ages = non_smoker_df.age
    estimated_charges = estimate_charges(ages, w, b)
    target_charges = non_smoker_df.charges
    sns.lineplot(x = ages, y = estimated_charges, color = "r")
    sns.scatterplot(x = non_smoker_df.age, y = target_charges, color = "b",alpha = 0.6)
    loss = rmse(target_charges, estimated_charges)
    print("RMSE loss: ", loss)
```

```python
try_parameters(220, 0)
```

```python
smoker_code = {"no":0,"yes":1}
df["smoker_code"] = df.smoker.map(smoker_code)

region_code = {"southeast":0,"southwest":1,
                           "northwest":2,"northeast":3}
df["region"] = df.region.map(region_code)

sex_code["sex"] = {"male":0,"female":1}
df.sex = df.sex.map(sex_code)
df.head()
```

```python
inputs, targets = smoker_df[["age", "bmi","children"]], smoker_df["charges"]

model = LinearRegression().fit(inputs, targets)
predictions = model.predict(inputs)
print(predictions, rmse(targets, predictions))
```
